```python
# stemming  -> mechanical process of reducing a word to its base form
# stemming is basically we remove inflection (prefix, suffix, tense) of the word. ex: (ly, ing, ed, s, es, etc)

# running > run
# undoable -> do
# studies > studi
# better > better
# happiness > happi

# ----------

# lematization -> linguistic process of reducing a word to its base form (lemma) which is a valid word in the language. 
# lematization is baisically we replace that word with its base word ex: (running -> run, unhappy -> sad, ran -> run, better -> good, etc)

# running > run
# studies > study
# better > good
# happiness > happy


# ---------------------------

# WordNet -> is a large lexical (language) database of English which has all the words of english. 
# and using this only we do lemmatization. 
# so when i say that lemmatization sees the dictionary for the base form of the word , it means that it looks into the wordnet corpus to find the base form of the word.

from nltk.stem import PorterStemmer, WordNetLemmatizer # import stemmer and lemmatizer
from nltk.corpus import wordnet # import the wordnet corpus which has information about words
# from nltk.book import *

ps = PorterStemmer()
lz = WordNetLemmatizer()
# class say object banaya and then we we do it like ps.method()...

word = 'running'
print(word)
print('stemming ->',ps.stem(word))
print('lematization ->',lz.lemmatize(word, pos='v'))

# stemming -> run
# lematization -> run

word = 'studying'
print(word)
print('stemming ->',ps.stem(word))
print('lematization ->',lz.lemmatize(word, pos='v'))

# stemming -> studi
# lematization -> study


# -------------------------
# stemming 
# -> mechanical shortening , Uses Algorithmic rules
# -> Acutal word may nahi latta hai - bas prefix , suffix hatata hai. (study -> studi)
# -> Fast

# lemmatization 
# -> linguistically correct base form , Sees the word in Dictionary for its base form . (so prefix, suffix, tense bhi haat jata hai)
# -> Acutal word lata hai (study -> study) 
# -> Slow

```

```python
# 1) lexical richness / lexical diversity --------------------------------

# trying to see ki -> total words may say kitne words duplicate hai. 
# using this 23 found the richness of the text.
# total is 44k , unique are 3k this means 41k words are duplicate.
# so the richness is very low .


######## Forumla :-
# lexical richness % = ( Total unique words / Total words ) * 100 

Lexical_Richness = len(set(text3)) / len(text3) * 100

print(Lexical_Richness)

# ex:-
# total words = 100
# unique words = 5
# so the richness is 5/100 * 100 = 5%

# lexical richness/diversity
# = 5/100 = 0.05
# = 5%


# 2) Average Occurrence of a Unique Word ---------------------------------

# tells:
# on average,
# how many times each unique word appears. 


######## Formula :-
# Average Repetition =  Total words / Total unique words

Average_Repetition = len(text3) / len(set(text3))

print(Average_Repetition)

# ex:-
# total words = 100
# unique words = 5
# so the average number of times a single unique word repeats is 100/5 = 20



# 3) Average Occurrence of a specific word ---------------------------------

# the average number of times the word 'the' repeats itself

###### Formula :-
# Average Occurrence of a specific word = ( Total Count of that word / Total words ) * 100

text3.count('the') / len(text3) * 100

# 'the' in text4
text4.count('the') / len(text4) * 100

# ex:-
# total words = 100
# total 'the' = 10
# so the average number of times the word 'the' repeats itself is 10/100 * 100 = 10%
# So in the Total words , 10% words are 'the'.


# Dont confuse :-
# Average occurrence - 2 chiz ka count kar skte hai -
# -> Unique words (because unique word he bar bar occur hota hai, duplicate word repeat nahi hota ...its just a repetation of a unique word only)
# -> Specific word (because specific word bhi bar bar occur hota hai, so we can count the average occurrence of a specific word also)
```

```python
# function for lexical richness / lexical diversity
# Basically tells the percentage of unique words in the text.
# if all words are unique then 100% 
def lexical_diversity(text):
  return len(set(text)) / len(text) * 100

# function for Average Occurrence of a Letter 
def average_letter_occurrence(text, letter):
  return text.count(letter) / len(text) * 100
```

```python
# collocations
# -----------------------------
# a collocation is a sequence of words that occur together more often than by chance.
# Words which are used together (New York , Heavy Rain, Strong Coffee, etc)
# so basically a collocation is a sequence of words that are commonly found together in the text. like "strong tea" is a collocation because "strong" and "tea" are commonly found together in the text. but "powerful tea" is not a collocation because "powerful" and "tea" are not commonly found together in the text.

# collocations cab be bigrams, trigrams, or n-grams. but bigrams are the most common type of collocations.
# ex:-
# strong tea   # 2 words together
# all the best   # 3 words together
# wish you good luck   # 4 words together

from nltk.book import *
# using .collocations() method we can find the collocations in the text.
text4.collocations()        # collocations are words which are used together . United states , strong coffee , heavy rain , nice shirt

# -> 
# United States; fellow citizens; years ago; four years; Federal
# Government; General Government; Vice President; American people; God
# bless; Chief Justice; one another; fellow Americans; Old World;
# Almighty God; Fellow citizens; Chief Magistrate; every citizen; Indian
# tribes; public debt; foreign nations







# n-grams 
# -------------------------
# so n-grams are basically n pair of words 
# ex - bigrams are pair of 2 words
# trigrams are pair of 3 words
# and so on...
# they are used for many purposes like Collocations finding, language modeling, text generation, etc.


# 1) bigrams
# ------------------------------
# bigrams are pair of 2 words 

# code :-
from nltk import bigrams
text = ['more','is','said','than','done']
list(bigrams(text))
# so it will give you the list of bigrams in the text. like 
# [('more', 'is'), ('is', 'said'), ('said', 'than'), ('than', 'done')]

# so using the bigrams we can find collocations in the text. 
# ex:- if the bigram "strong tea" occurs more times then we can say that "strong tea" is a collocation in the text. 


# 2) trigrams
# ------------------------------
# trigrams are pair of 3 words
from nltk import trigrams
text = ['more','is','said','than','done']
list(trigrams(text))
# so it will give you the list of trigrams in the text. like
# [('more', 'is', 'said'), ('is', 'said', 'than'), ('said', 'than', 'done')]
```

```python

# frequency distribution using FreqDist in NLTK

# -------------------------------------------------------

# it gives the count of each word in the text.
# and count is just the number of times a word occurs in the text.
# so frequency distribution is just a dictionary with words as keys and their frequencies as values.
# [(word1, count1), (word2, count2), ...]

# use FreqDist() function and pass the text -> to make the frequency distribution of the text.
# which is basically a dictionary with words as keys and their frequencies as values.
# then we can use the methods of FreqDist to get the frequency of each word, most common words, etc


# Frequency of all the words in the text1
fdist1 = FreqDist(text1) # shows the number of times each words occurs
print(fdist1)

# -> FreqDist({',': 18713, 'the': 13721, '.': 6862, 'of': 6536, 'and': 6024, 'a': 4569, 'to': 4542, ';': 4072, 'in': 3916, 'that': 2982, ...})


# Frequency of a specific word
fdist1['whale']
# -> 906


# Frequency of the top most common/repeated words
# using .most_common(top_N) method 
fdist1.most_common(50)   # top 50 most repeated/frequent words
# -> [(',', 18713), ('the', 13721), ('.', 6862), ('of', 6536), ('and', 6024), ('a', 4569), ('to', 4542), (';', 4072), ('in', 3916), ('that', 2982), ...]

```

```python
# fine-grained selection of words
# -----------------------------------
# basically selecting some specific words out of total words -> which fulfills the property / condition
# 

# ex:-
selected_words = []
Vocabulary = set(text1)  # all the unique words in text1 # other wise all duplicate words will also be selected.
for word in Vocabulary:
  if len(word) > 5:   # condition / property
    selected_words.append(word)


# its basically :-

# word = [] # making a empty list

# for w in V:
#   if len(w) > 15:
#     word = word + [w]   # we can also simply append in the word list.

# print(word)


# Formula :-
# {w | w ε V & P(w)}          
# V means - Voculabouary (all unique words)   # ε means - belong (epselone)
# the set of all words
# words belong to the vocabulary
# and word has the property


# {w | w ε V & P(w)}
# 1) w ε V  => a word in vocabulary
# 2) P(w) => and that word fulfills the property/condition
# 3) then w is selected in the fine-grained selection of words. 



# so we can do it in python using list comprehension
# selected_words = [w for w in V if p(w)]


V = set(text1) # list of all unique words in text1 (vocabulary)

long_words = [w for w in V if len(w) > 15]  # list comphrehension
# so here - we loop into Vocabulary list and see if that word is 15 char long, and then if yes then we return that w .
#  and fir vo word will be appended in the long_word list.  and it iterates and checks and soo onn.


# 1) for w in V -> loop over 
# 2) if len(w) > 15 -> condition / property
# 3) w -> then return that word if it fulfills the condition
# 4) long_words = [] -> append that returned word to the list

print(long_words)



# now we sort the words
print(sorted(long_words))
```

```python
# fdist1 |= fdist2
# update fdist1 with counts from fdist2

# 

fdist1 = FreqDist(text1)
# so fdist1 is the frequency distribution of text1
fdist2 = FreqDist(text2)
# so fdist2 is the frequency distribution of text2

fdist1 |= fdist2
# update fdist1 with counts from fdist2 (so it will add the counts of each word in fdist2 to the counts of the same word in fdist1, if the word is not present in fdist1 then it will add that word to fdist1 with its count from fdist2)
# basically merge 2 dictonaries dict_1 = {word1: count1, word2: count2, ...} and dict_2 = {word1: count1, word2: count2, ...} to get merge_dict = {word1: count1 + count1, word2: count2 + count2, ...}

fdist1
# will have both words of text1 and text2 with their counts updated 
```

```python
# Other Lexical Calculations :-

# ----------------------

# Textual Data

Text = "Call me Ishmael. The family of Dashwood had long been settled in Sussex."

# ----------------------

# Tokenize the text into words (split by space)

Text = Text.split()  # so it will split the text into words based on space 
# Text = ['Call', 'me', 'Ishmael', '.' , 'The', 'family', 'of', 'Dashwood', 'had', 'long', 'been', 'settled', 'in', 'Sussex', '.']

# ----------------------

# Total number of characters = sum of the length of each word in the text (leaving spaces)

Total_Number_of_characters = sum(len(word) for word in Text)
print (Total_Number_of_characters) 
# output 70 

# ----------------------
 
# Total number of words = length of the list of words in the text 

Total_Number_of_words = len(Text)
print(Total_Number_of_words)
# output 15 (since there are 15 words in the text, which are 'Call', 'me', 'Ishmael', '.', 'The', 'family', 'of', 'Dashwood', 'had', 'long', 'been', 'settled', 'in', 'Sussex', '.')

# ----------------------

# Total number of vocabulary = length of the set of words in the text 

Total_Number_of_Vocabulary = len(set(word for word in Text))
print(Total_Number_of_Vocabulary)
# output 14 (since there are 14 unique words in the text, which are 'Call', 'me', 'Ishmael', '.', 'The', 'family', 'of', 'Dashwood', 'had', 'long', 'been', 'settled', 'in', 'Sussex')

# ----------------------

# Total number of sentences = count of the number of sentences in the text

Text2 = [sentence for sentence in Text if sentence.endswith('.')] # so it will give you the list of sentences in the text, which are the words that end with a period '.'
Total_Number_of_sentences = len(Text2)
print(Total_Number_of_sentences)
# output 2 (since there are 2 sentences in the text, which are 'Call me Ishmael.' and 'The family of Dashwood had long been settled in Sussex.')

# ----------------------

# Average number of characters per word = Total number of characters / Total number of words

Average_Number_of_characters_per_word = Total_Number_of_characters / Total_Number_of_words
print(Average_Number_of_characters_per_word)
# output 4.666666666666667 (since there are 70 characters in the text and 15 words in the text, so the average number of characters per word is 70/15 = 4.66)

# ----------------------

# Average number of words per sentence = Total number of words / Total number of sentences

Average_Number_of_words_per_sentence = Total_Number_of_words / Total_Number_of_sentences
print(Average_Number_of_words_per_sentence)
# output 1.0 (since there are 15 words in the text and 15 sentences in the text, so the average number of words per sentence is 15/15 = 1.0)

# ----------------------

# Lexical Diversity = Total Number of Vocabulary / Total Number of words * 100

Lexcical_Diversity = Total_Number_of_Vocabulary / Total_Number_of_words * 100
print(Lexcical_Diversity)
# output 93.33333333333333 (since there are 14 unique words in the text and total number of words in the text is 15, so the lexical diversity is 14/15 * 100 = 93.33%)

# ----------------------

# Average Occurrence of a Unique Word = Total Number of words / Total Number of Vocabulary

Average_Occurrence_of_a_Unique_Word = Total_Number_of_words / Total_Number_of_Vocabulary
print(Average_Occurrence_of_a_Unique_Word)
# output 1.0714285714285714 (since there are 15 words in the text and 14 unique words in the text, so the average occurrence of a unique word is 15/14 = 1.07)

# ----------------------

# Average Occurrence of a Specific Word = ( Total Count of that word / Total words ) * 100
  
Average_Occurrence_of_a_Specific_Word = Text.count('the') / Total_Number_of_words * 100
print(Average_Occurrence_of_a_Specific_Word)
# output 13.333333333333334 (since the word 'the' occurs 2 times in the text and total number of words in the text is 15, so the average occurrence of the word 'the' is (2/15) * 100 = 13.33%)


```

```python
# Assignment 4: Conditional Frequency Distribution
from nltk.corpus import brown
from nltk import ConditionalFreqDist

# Create Conditional Frequency Distribution
cfd = ConditionalFreqDist(
    (genre, word.lower())
    for genre in ['romance', 'humor', 'mystery']
    for word in brown.words(categories=genre)
    if word.lower() in ['love', 'hate', 'fear']
)

# Tabulate results
cfd.tabulate(conditions=['romance', 'humor', 'mystery'], samples=['love', 'hate', 'fear'])
```

```python

# ============================================
# COMBINE DATA INTO SINGLE TEXT
# ============================================

text = " ".join(documents)

print(text)

```

```python

# ============================================
# SENTENCE TOKENIZATION
# ============================================

sentences = sent_tokenize(text)

print(sentences)

```

```python

# ============================================
# WORD TOKENIZATION
# ============================================

words = word_tokenize(text)

print(words)

```

```python

# ============================================
# LOWERCASE
# ============================================

words_lower = [word.lower() for word in words]

print(words_lower)

```

```python

# ============================================
# REMOVE NON ALPHABETIC WORDS
# ============================================

words_clean = [
    word
    for word in words_lower
    if word.isalpha()
]

print(words_clean)

```

```python

# ============================================
# REMOVE STOPWORDS
# ============================================

stop_words = set(stopwords.words("english"))

filtered_words = [

    word

    for word in words_clean

    if word not in stop_words

]

print(filtered_words)

```

```python

# ============================================
# VERY SIMPLE TEXT CLEANING FUNCTION
# ============================================

def clean_text(text):

    text = text.lower()

    words = word_tokenize(text)

    words = [

        word

        for word in words

        if word.isalpha()

    ]

    words = [

        word

        for word in words

        if word not in stop_words

    ]

    return words


print(clean_text(documents[0]))
```
